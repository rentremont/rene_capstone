```python
disneylist = pd.read_csv('data/mylist_lmpssa.csv')
disneylist
```

```python
disneylist=disneylist.drop(['Unnamed: 0'], axis=1)
disneylist
```

```python
disneylist=disneylist.rename(columns={'Release':'title','category':'category', 'To Date/Lifetime Gross in US and Canada': 'lifetime_gross_us_ca','Release Date_x':'release_date','Max Theaters_x':'max_theaters','Days':'total_days','Opening_x':'opening_weekend','Open Th_x':'opening_theaters','Distributor':'distributor'})
disneylist
```

```python
disneylist['release_date'] = pd.to_datetime(disneylist['release_date'], format = '%d-%b-%y')
```

```python
avg_ticket_per_year=pd.read_csv('data/avg_ticket_per_year.csv')
avg_ticket_per_year
```

```python
avg_ticket_per_year['inflation_rate']= (10.53-avg_ticket_per_year['avg_price'])/avg_ticket_per_year['avg_price']
```

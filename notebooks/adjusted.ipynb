### Webscraping to get the lifetime gross adjusted for inflation

```python
import requests
import pandas as pd
```

```python
URL = 'https://www.boxofficemojo.com/chart/top_lifetime_gross_adjusted/?adjust_gross_to=2022'

response = requests.get(URL)
```

```python
response.status_code
```

```python
response.text
```

```python
from bs4 import BeautifulSoup as BS
```

```python
soup = BS(response.text)
soup
```

```python
soup.findAll('table')
```

```python
tables=pd.read_html('https://www.boxofficemojo.com/chart/top_lifetime_gross_adjusted/?adjust_gross_to=2022')
tables
```

```python
len(tables)
```

```python
adjusted_tables1=tables[0]
adjusted_tables1
```

```python
tables2=pd.read_html('https://www.boxofficemojo.com/chart/top_lifetime_gross_adjusted/?adjust_gross_to=2022&offset=200')
tables2
```

```python
adjusted_tables2=tables2[0]
adjusted_tables2
```

```python
tables3=pd.read_html('https://www.boxofficemojo.com/chart/top_lifetime_gross_adjusted/?adjust_gross_to=2022&offset=400')
tables3
```

```python
adjusted_tables3=tables3[0]
adjusted_tables3
```

```python
tables4=pd.read_html('https://www.boxofficemojo.com/chart/top_lifetime_gross_adjusted/?adjust_gross_to=2022&offset=600')
tables4
```

```python
adjusted_tables4=tables4[0]
adjusted_tables4
```

```python
tables5=pd.read_html('https://www.boxofficemojo.com/chart/top_lifetime_gross_adjusted/?adjust_gross_to=2022&offset=800')
tables5
```

```python
adjusted_tables5=tables5[0]
adjusted_tables5
```

```python
adjusted_data = pd.concat([adjusted_tables1, adjusted_tables2, adjusted_tables3, adjusted_tables4, adjusted_tables5])
adjusted_data
```

import os  
os.makedirs('data', exist_ok=True)  
adjusted_data.to_csv('data/adjusted.csv')

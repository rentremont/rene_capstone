```python
URL = 'https://pro.imdb.com/company/co0226183/boxoffice/?view=releases&ref_=mojo_bn_table_1&rf=mojo_bn_table_1'

response = requests.get(URL)
```

```python
titles = soup.findAll('h3', attrs={'class':'lister-item-header'})
titles_list = []

for x in  titles:
    titles_list.append(x.text)
```

```python
URL = 'https://www.imdb.com/list/ls033609554/?st_dt=&mode=detail&'
titles_list2 = []
for page in range (1,7):
    req = requests.get(URL + 'page=' + str(page) + '&ref_=ttls_vm_dtl&sort=release_date,asc')
    soup = BS(req.text, 'html.parser')
    title = soup.findAll('h3', attrs={'class':'lister-item-header'})
    for x in title:
        titles_list2.append(x.text)
titles_list2
```

```python
titles_listfull=pd.DataFrame(titles_list2)
titles_listfull
```

```python
titles_listfull.to_csv()
```
